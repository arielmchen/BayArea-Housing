```python
#conduct anova test including dummy variables
%matplotlib inline

# load packages
import scipy.stats as stats

# stats f_oneway functions takes the groups as input and returns ANOVA F and p value
fvalue, pvalue = stats.f_oneway(zipcode_dummies['Zipcode_940'], zipcode_dummies_all['Zipcode_941'], zipcode_dummies_all['Zipcode_943'], zipcode_dummies_all['Zipcode_944'],  zipcode_dummies_all['Zipcode_945'], zipcode_dummies_all['Zipcode_946'], zipcode_dummies_all['Zipcode_947'], zipcode_dummies_all['Zipcode_949'], zipcode_dummies_all['Zipcode_950'], zipcode_dummies_all['Zipcode_951'], zipcode_dummies_all['Price'], zipcode_dummies_all['Living Area'])
print('F statistic = {:5.3f} and probability p = {:5.3f}'.format(fvalue, pvalue)) 
```
